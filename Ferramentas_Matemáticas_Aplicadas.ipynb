```python
##############################################
# Máximos e Mínimos de funções com derivadas #
##############################################
 
# Exemplo 1: relação entre preço de venda "x" e o lucro "y" de um produto
# Qual o preço de venda que maximiza o lucro? E qual o lucro máximo?

from sympy import Eq, diff, solve, symbols 

x,y=symbols('x y')
y=-4*x**2+4000*x-200000
df=diff(y, x)
d2f=diff(y, x, 2)

# Encontrando o preço ótimo de venda (faço df=0 e resolvo a equação)
p=solve(Eq(df, 0))

# Encontrando o lucro máximo (substituo "p" no valor de "x" na função "y")
l=y.subs(x, p[0])

# Teste da derivada segunda (avaliar se o ponto crítico é um ponto de máxima
# f''(c)<0 ; ponto de mínima f''(c)>0 ; ou um ponto de inflexão horizontal 
# f''(c)=0 ou f''n(c)!=0 se n ímpar)
ds=d2f.subs(x,p[0])

print('Preço ótimo: ', p[0])
print('Lucro máximo: ', l)
print('Derivada segunda: ', ds)
```

```python
# Exemplo 2: custo "c" referente à produção diária de "x" unidades
# Qual é o nível de produção que minimiza o custo?

from sympy import symbols, diff, Eq, solve
import matplotlib.pyplot as plt
import numpy as np

x, c = symbols('x c')
c = x**2-20*x+300
df = diff(c, x)
d2f = diff(c, x, 2)
p = solve(Eq(df, 0)) # produção ótima
ds = d2f.subs(x, p[0]) # teste da derivada segunda

print('Produção ótima: ', p[0])
print('Derivada segunda: ', ds)

# Gráfico

x = np.linspace(0, 25, 100)
c = x**2-20*x+300
plt.plot(x, c)
plt.show()
```

```python
# Exemplo 5: um objeto se desloca em linha reta, e a relação entre a distância
# considerada em metros do objeto à origem e o tempoem segundos é dada por
# s = 2*t**2 + 3*t
# Savendo que a velocidade corresponde À derivada de "s" em relação a "t",
# determine a vellocidade do objeto quando t = 2 segundos

from sympy import diff, solve, Eq, symbols

s, t = symbols('s t')
s = 2*t**2 + 3*t
ds = diff(s, t)

v = ds.subs(t, 2)

print('Velocidade: %.2f m/s' %v)

```

```python
################################
# Otimização em 3D             #
# Cálculo com várias variáveis #
################################

# Exemplo 1: dada a função f(x, y) = x**2 + y**2, faça o gráfico e em seguida
# obtenha e classifique os pontos críticos

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

xx = np.linspace(-5, 5, 100)
yy = np.linspace(-5, 5, 100)
x, y = np.meshgrid(xx, yy)
f = x**2 + y**2
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, f)

# Cálculo dos PONTOS CRÍTICOS

from sympy import *

x, y, f = symbols('x y f')
f = x**2 + y**2

# Para DETERMINAR os pontos críticos utiliza-se a derivada primeira
fx = diff(f, x)
fy = diff(f, y)

px = solve(Eq(fx, 0))
py = solve(Eq(fy, 0))

# Para CLASSIFICAR os pontos críticos utiliza-se a derivada segunda
fxx = diff(f, x, 2)
fyy = diff(f, y, 2)
fxy = diff(fx, y)
fyx = diff(fy, x)

fxxp = fxx.subs({x: px[0], y: py[0]})
fyyp = fyy.subs({x: px[0], y: py[0]})
fxyp = fxy.subs({x: px[0], y: py[0]})
fyxp = fyx.subs({x: px[0], y: py[0]})

# Cálculo do determinante da matriz formada pelas quatro derivadas parciais
# de segunda ordem (Matriz Hessiana)
D = fxxp*fyyp - fxyp*fxyp

print('Solução: (%.2f, %.2f)' %(px[0], py[0]))
print('Determinante: ', D)
print('Derivada segunda em relação a x: ', fxxp)
```

```python
# Considerando a função f(x,y) = (1 - x)**2 + 2*(3 - y)**2
# faça o gráfico e em seguida determine e classifique os pontos críticos

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

xx = np.linspace(-5, 5, 100)
yy = np.linspace(-5, 5, 100)
x, y = np.meshgrid(xx, yy)
f = (1 - x)**2 + 2*(3 - y)**2
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, f)

from sympy import symbols, diff, Eq, solve

x, y, f = symbols('x y f')
f = (1 - x)**2 + 2*(3 - y)**2

fx = diff(f, x)
fy = diff(f, y)

fxx = diff(f, x, 2)
fxy = diff(fx, y)
fyx = diff(fy, x)
fyy = diff(f, y, 2)

px = solve(Eq(fx, 0))
py = solve(Eq(fy, 0))

fxxp = fxx.subs({x: px[0], y: py[0]})
fxyp = fxy.subs({x: px[0], y: py[0]})
fyxp = fyx.subs({x: px[0], y: py[0]})
fyyp = fyy.subs({x: px[0], y: py[0]})

D = fxxp*fyyp - fxyp*fyxp

print('Solução: (%.2f, %.2f)' %(px[0], py[0]))
print('Determinante hessiano: ', D)
print('Derivada segunda em relação a x: ', fxxp)
```

```python
# Exemplo 2: área abaixo da curva "y=x³", de x=1 a x=3

from sympy import symbols, integrate

x, y = symbols('x y')
y = x**3
A = integrate(y, (x, 1, 3))

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.5, 3.5, 1000)
y = x**3
plt.plot(x, y, color='blue')
plt.axhline(color='blue')
plt.fill_between(x, y, where=[(x > 1) and (x < 3) for x in x], color='magenta')
print('Área: ', A)
```

```python
# Exemplo 3: área limitada pelo gráfico da função "y" e pelo eixo "x"

import matplotlib.pyplot as plt
import numpy as np
from sympy import *

x, y = symbols('x y')
y = -x**2 + 4*x + 1
coeff=[-1, 4, 1]
r = np.roots(coeff)
A = integrate(y, (x, min(r), max(r)))

x = np.linspace(min(r) - 0.5, max(r) + 0.5, 1000)
y = -x**2 + 4*x + 1
plt.plot(x, y, color='blue')
plt.axhline(color='blue')
plt.fill_between(x, y, where=[(x > min(r)) and 
                              (x < max(r)) for x in x], color='yellow')

print('Área: ', A)
```
