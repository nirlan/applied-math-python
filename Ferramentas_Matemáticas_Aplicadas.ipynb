```python
##############################################
# Máximos e Mínimos de funções com derivadas #
##############################################
 
# Exemplo 1: relação entre preço de venda "x" e o lucro "y" de um produto
# Qual o preço de venda que maximiza o lucro? E qual o lucro máximo?

from sympy import Eq, diff, solve, symbols 

x,y=symbols('x y')
y=-4*x**2+4000*x-200000
df=diff(y, x)
d2f=diff(y, x, 2)

# Encontrando o preço ótimo de venda (faço df=0 e resolvo a equação)
p=solve(Eq(df, 0))

# Encontrando o lucro máximo (substituo "p" no valor de "x" na função "y")
l=y.subs(x, p[0])

# Teste da derivada segunda (avaliar se o ponto crítico é um ponto de máxima
# f''(c)<0 ; ponto de mínima f''(c)>0 ; ou um ponto de inflexão horizontal 
# f''(c)=0 ou f''n(c)!=0 se n ímpar)
ds=d2f.subs(x,p[0])

print('Preço ótimo: ', p[0])
print('Lucro máximo: ', l)
print('Derivada segunda: ', ds)
```

```python
################################
# Otimização em 3D             #
# Cálculo com várias variáveis #
################################

# Exemplo 1: dada a função f(x, y) = x**2 + y**2, faça o gráfico e em seguida
# obtenha e classifique os pontos críticos

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

xx = np.linspace(-5, 5, 100)
yy = np.linspace(-5, 5, 100)
x, y = np.meshgrid(xx, yy)
f = x**2 + y**2
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, f)

# Cálculo dos PONTOS CRÍTICOS

from sympy import *

x, y, f = symbols('x y f')
f = x**2 + y**2

# Para DETERMINAR os pontos críticos utiliza-se a derivada primeira
fx = diff(f, x)
fy = diff(f, y)

px = solve(Eq(fx, 0))
py = solve(Eq(fy, 0))

# Para CLASSIFICAR os pontos críticos utiliza-se a derivada segunda
fxx = diff(f, x, 2)
fyy = diff(f, y, 2)
fxy = diff(fx, y)
fyx = diff(fy, x)

fxxp = fxx.subs({x: px[0], y: py[0]})
fyyp = fyy.subs({x: px[0], y: py[0]})
fxyp = fxy.subs({x: px[0], y: py[0]})
fyxp = fyx.subs({x: px[0], y: py[0]})

# Cálculo do determinante da matriz formada pelas quatro derivadas parciais
# de segunda ordem (Matriz Hessiana)
D = fxxp*fyyp - fxyp*fxyp

print('Solução: (%.2f, %.2f)' %(px[0], py[0]))
print('Determinante: ', D)
print('Derivada segunda em relação a x: ', fxxp)
```

```python
# Considerando a função f(x,y) = (1 - x)**2 + 2*(3 - y)**2
# faça o gráfico e em seguida determine e classifique os pontos críticos

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

xx = np.linspace(-5, 5, 100)
yy = np.linspace(-5, 5, 100)
x, y = np.meshgrid(xx, yy)
f = (1 - x)**2 + 2*(3 - y)**2
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, f)

from sympy import symbols, diff, Eq, solve

x, y, f = symbols('x y f')
f = (1 - x)**2 + 2*(3 - y)**2

fx = diff(f, x)
fy = diff(f, y)

fxx = diff(f, x, 2)
fxy = diff(fx, y)
fyx = diff(fy, x)
fyy = diff(f, y, 2)

px = solve(Eq(fx, 0))
py = solve(Eq(fy, 0))

fxxp = fxx.subs({x: px[0], y: py[0]})
fxyp = fxy.subs({x: px[0], y: py[0]})
fyxp = fyx.subs({x: px[0], y: py[0]})
fyyp = fyy.subs({x: px[0], y: py[0]})

D = fxxp*fyyp - fxyp*fyxp

print('Solução: (%.2f, %.2f)' %(px[0], py[0]))
print('Determinante hessiano: ', D)
print('Derivada segunda em relação a x: ', fxxp)
```

```python
# Exemplo 2: vetor u = (3, -5, 2) e v = (4, 6, 3),
# calcule u*v (PRODUTO ESCALAR ou produto interno)
# e uXv (PRODUTO VETORIAL ou produto externo)

import numpy as np

u = np.array([[3, -5, 2]])
v = np.array([[4, 6, 3]])

# produto escalar
uv = np.inner(u, v)

# produto vetorial
uXv = np.cross(u, v)

print(uv)
print(uXv)
```

```python
#####################
# SISTEMAS LINEARES #
#####################

# Exemplo 1:

# {4x + 2y - z = 7
# {3x + 3y + 2z = 20
# {5y + 2z = -1

import numpy as np

A = np.array([[4, 2, -1], [3, 3, 2], [0, 5, 2]])
b = np.array([[7], [20], [-1]])
x = np.linalg.solve(A, b)

print(x)
```

```python
# Exemplo 2: MATRIZ SINGULAR (infinitas soluções ou não tem solução)

# {x + 3y + 4z = 18
# {2x - y + z = 10
# {-4y + 2y - 2z = -7

import numpy as np

A = np.array([[1, 3, 4], [2, -1, 1], [-4, 2, -2]])
b = np.array([[18], [10], [-7]])
x = np.linalg.solve(A, b)

print(x)
```

```python
# Exemplo 3:

# Em uma transportadora, 10 pallets com 12 caixas pesam 844 Kg.
# Nas mesmas condições, 8 pallets com 10 caixas cada um deles
# pesam 576 Kg. Qual é o peso de cada pallet e qual é o peso de cada caixa?

import numpy as np

A = np.array([[10, 120], [8, 80]])
b = np.array([[844], [576]])
x = np.linalg.solve(A, b)

print('Cada pallet pesa %.2f kg e cada caixa pesa %.2f Kg' %(x[0, 0], x[1, 0]))
```

```python
########################################################
# RESOLUÇÃO DE SISTEMAS LINEARES COM NÚMEROS COMPLEXOS #
########################################################

# Exemplo 1: número complexo na forma z = a + b*i

#{(2 + j)a + (5 - j)b = 3 + 8j
#{(5 - j)a + (7 + 4j)b = 2 + 5j

import numpy as np

A = np.array([[complex(2, 1), complex(5, -1)],
              [complex(5, -1), complex(7, 4)]])
b = np.array([[complex(3, 8)], 
              [complex(2, 5)]])
np.linalg.solve(A, b)
```

```python
# Exemplo 2: número complexo na forma polar z = r*(cosø + i*senø)

#{(3 + 2j)a + (-2 - 6j)b = 70 <0º
#{(-2 - 6j)a + (10 + j)b = 110 <30º

# Temos que transformar a forma polar na forma "a + b*i"

import numpy as np

A = np.array([[complex(3, 2), complex(-2, -6)],
              [complex(-2, -6), complex(10, 1)]])

x1 = np.cos(np.deg2rad(0)) #parte real de 70 <0º
y1 = np.sin(np.deg2rad(0)) #parte imaginária de 70 <0º
x2 = np.cos(np.deg2rad(30)) #parte real de 110 <30º
y2 = np.sin(np.deg2rad(30)) #parte imaginária de 110 <30º

# 70 e 110 correspondem ao raio "r" --> z = r*(cosø + i*senø) 
b = np.array([[70*complex(x1, y1)],
              [110*complex(x2, y2)]])

np.linalg.solve(A, b)
```

```python
####################
# REGRESSÃO LINEAR #
####################

# Exemplo 1:
# Em uma indústria, a produção nos quatro primeiros meses do ano foi,
# respectivamente, 5000, 5300, 5200 e 5500 unidades. Obtenha a reta de 
# regressão, o coeficiente de correlação e faça o gráfico contendo os dados 
# originais e a reta de regressão.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.array([1, 2, 3, 4])
y = np.array([5000, 5300, 5200, 5500])
a, b, correlacao, p, erro = stats.linregress(x, y)
print('Reta de regressão: y = %.2fx + %.2f' % (a, b))
print('Coeficiente de correlação de Pearson: r = %.2f' % correlacao)
print('p = %.2f' % p)
print('Erro = %.2f' % erro)

plt.plot(x, y, 'o', label='Dados originais')
f = a*x + b
plt.plot(x, f, 'r', label='Reta de regressão')
plt.ylim(4900, 5600)
plt.legend()
plt.show()
```

```python
################
# INTERPOLAÇÃO #
################

# Exemplo 1:
# Obtenha graficamente a parábola relacionada à estrutura de uma ponte que tem
# 20 metros de comprimento e 7 metros de altura

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import *

x = [0, 10, 20]
y = [0, 7, 0]
f = interp1d(x, y, kind='quadratic')
xi = np.linspace(1, 20, 200)
yi = f(xi)
plt.plot(x, y, 'o')
plt.plot(xi, yi)
plt.show()
```

```python
################################################
# PROBLEMAS DE PROGRAMAÇÃO LINEAR / OTIMIZAÇÃO #
################################################

# Exemplo 1:

# Uma importadora pretende comprar pen drives, caixas de som e 
# carregadores para celular. Essa importadora tem 30 mil dólares
# e pretende decidir quantas unidades de cada um desses itens
# irá adquirir de modo a maximizar o lucro referente à posterior
# venda dessas mercadorias:

# ITEM            Preço de custo   Lucro unitário   Qtd mínima    Qtd máxima
# Caixa de som    $ 5.90           $ 7.10           50            200
# Carregador      $ 2.38           $ 2.52           200           -
# Pen drive       $ 2.90           $ 2.00           100           -

# Resolução -->

# VARIÁVEIS: 
#                   x1 = qtd de caixas de som
#                   x2 = qtd carregadores para celular
#                   x3 = qtd pen drives

# FUNÇÃO OBJETIVO: 
#                   LucroMax = 7.10*x1 + 2.52*x2 + 2.00*x3

# RESTRIÇÕES: 
#                   5.90*x1 + 2.38*x2 + 2.90*x3 <= 30000
#                   50 <= x1 <= 200
#                   x2 >= 200
#                   x3 >= 100

# FORMULAÇÃO:
#             Lmax = 7.10*x1 + 2.52*x2 + 2.00*x3
#             5.90*x1 + 2.38*x2 + 2.90*x3 <= 30000
#             x1 >= 50
#             x1 <= 200
#             x2 >= 200
#             x3 >= 100

import sys
!{sys.executable} -m pip install pulp
from pulp import *

prob = LpProblem('Exemplo1', LpMaximize)
x1 = LpVariable('Caixa de som', 0)
x2 = LpVariable('Carregador', 0)
x3 = LpVariable('Pen drive', 0)
prob += 7.1*x1 + 2.52*x2 + 2*x3
prob += 5.9*x1 + 2.38*x2 + 2.9*x3 <= 30000
prob += x1 >= 50
prob += x1 <= 200
prob += x2 >= 200
prob += x3 >= 100
prob.solve()
for v in prob.variables():
  print(v.name, '=', v.varValue)
print('Lucro máximo = ', value(prob.objective))
```
